```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
import matplotlib.pyplot as plt
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import f_classif
```

```python
data = pd.read_csv("drug200.csv")
```

```python
dataset = data.values
dataset
```

```python
X = dataset[:, :-1]
X
```

```python
Y = dataset[:, -1]
Y
```

```python
X_new = pd.DataFrame(X)
y_new = pd.DataFrame(Y)
```

```python
X_new = X_new.drop([0, 4], axis=1)
X_new
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, test_size=0.33, random_state=42)
```

```python
print(f"The Train shape of X : {X_train.shape} and y: {y_train.shape}")
print(f"The Test shape of X: {X_test.shape} and y: {y_test.shape}")
```

```python
def prepare_inputs(X_train, X_test):
    ordinal_encoder = OrdinalEncoder()
    ordinal_encoder.fit(X_train)
    X_train_enc = ordinal_encoder.transform(X_train)
    X_test_enc = ordinal_encoder.transform(X_test)
    return X_train_enc, X_test_enc
```

```python
def prepare_target(y_train, y_test):
    label_encoder = LabelEncoder()
    label_encoder.fit(y_train)
    y_train_enc = label_encoder.transform(y_train)
    y_test_enc = label_encoder.transform(y_test)
    return y_train_enc, y_test_enc
```

```python
X_train_enc, X_test_enc = prepare_inputs(X_train, X_test)
```

```python
y_train_enc, y_test_enc = prepare_inputs(y_train, y_test)
```

### ANOVA F_statistic score based feature selection

```python
def select_features(X_train,y_train,X_test):
    fs=SelectKBest(score_func=f_classif,k=3)
    fs.fit(X_train,y_train)
    X_train_fs=fs.transform(X_train)
    X_test_fs=fs.transform(X_test)
    return X_train_fs,X_test_fs,fs
```

```python
#Selecting best features and plotting
X_train_fs,X_test_fs,fs=select_features(X_train_enc,y_train_enc,X_test_enc)
plt.bar([i for i in range(len(fs.scores_))],fs.scores_)
plt.show()
```

```python
#feature scores
fs.scores_
```

### Mutual information based feature selection

```python
def select_features_2(X_train,y_train,X_test):
    fs=SelectKBest(score_func=mutual_info_classif,k=3)
    fs.fit(X_train,y_train)
    X_train_fs=fs.transform(X_train)
    X_test_fs=fs.transform(X_test)
    return X_train_fs,X_test_fs,fs
```

```python
#Selecting best features using MI and plotting
X_train_fs2,X_test_fs2,fs2=select_features_2(X_train_enc, y_train_enc, X_test_enc)
plt.bar([i for i in range(len(fs2.scores_))],fs2.scores_)
plt.show();
```

```python
#feature scores
fs2.scores_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model1=RandomForestClassifier()
model1.fit(X_train_enc,y_train_enc)
yhat=model1.predict(X_test_enc)
accuracy=accuracy_score(y_test_enc,yhat)
print("Accuracy: %.2f" %(accuracy*100))
```

```python
model3=RandomForestClassifier()
model3.fit(X_train_fs2,y_train_enc)
yhat=model1.predict(X_test_fs2)
accuracy=accuracy_score(y_test_enc,yhat)
print("Accuracy: %.2f" %(accuracy*100))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
```

```python
def load_dataset(fname):
    data=pd.read_csv(fname,header=None)
    dataset=data.values
    X=dataset[:,:-1]
    y=dataset[:,-1]
    return X,y
```

```python
X, y = load_dataset("drug200.csv")
```

```python
#define the evaluation method - k-fold cross validation & k=10
cv=RepeatedStratifiedKFold(n_splits=10,n_repeats=3,random_state=1)
```

```python
#define the pipeline to evaluate
model=RandomForestClassifier()
fs=SelectKBest(score_func=f_classif)
pipeline=Pipeline(steps=[('anova',fs),('lr',model)])
```

```python
#define the grid
grid=dict()
grid['anova__k']=[i+1 for i in range(X.shape[1])]
```

```python
#define the grid search
search=GridSearchCV(pipeline, grid, scoring='accuracy',n_jobs=-1, cv=cv)
results=search.fit(X_train_enc, y_train_enc)
```

```python
print('Best Mean Accuracy: %.3f' % results.best_score_)
print('Best config %s' %results.best_params_)
```

```python
#comparing different no. of features selected using ANOVA f-test
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score

def evaluate_model(model):
    cv=RepeatedStratifiedKFold(n_repeats=3,n_splits=10, random_state=1)
    scores=cross_val_score(model,X_test_enc,y_test_enc,scoring='accuracy',cv=cv,n_jobs=1)
    return scores
```

```python
#define the no. of features
num_features=[i+1 for i in range(X.shape[1])]
```

```python
#enumerate each no. of feature
results=list()
for k in num_features:
#create pipeline
    model=RandomForestClassifier()
    fs=SelectKBest(score_func=f_classif,k=3)
    pipeline=Pipeline(steps=[('anova',fs),('lr',model)])
```

```python
#evaluate the model
scores=evaluate_model(pipeline)
results.append(scores)
```

```python
print('#%d %.3f (%.3f)' %(k,mean(scores),std(scores)))
```
